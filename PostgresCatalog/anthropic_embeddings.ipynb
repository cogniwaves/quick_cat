```python
def create_embeddings(text):
    """Create embeddings for a list of food items."""
    # Add food context to help the model understand the domain
    contextualized_item = f"food item: {text}"
    # Generate embedding
    result = vo.embed(contextualized_item, model="voyage-3-large", input_type="document")
    
    
    return result.embeddings[0]
```

```python
from psycopg2 import sql

DB_URL = "postgresql://postgres.oqvdwtiwrzyjpnouwxch:f87JpR9Uvud6NR3HwbP@aws-0-ca-central-1.pooler.supabase.com:5432/postgres"

conn = psycopg2.connect(DB_URL)
cursor = conn.cursor()
cursor.execute("SET search_path TO catalog, public;")

embedding = create_embeddings('Épaule Boeuf')

# Convert list to PostgreSQL vector literal
# Example: "ARRAY[-0.06, 0.12, ...]::catalog.vector"
embedding_literal = "ARRAY" + str(embedding) + "::vector"

# Use psycopg2.sql to safely compose the query
query = sql.SQL(
    "SELECT subcategory_id, subcategory_name_en, voyage_1024 <-> {vec} AS distance "
    "FROM catalog.subcategory2 "
    "WHERE voyage_1024 IS NOT NULL "
    "ORDER BY distance "
    "LIMIT 10;"
).format(vec=sql.SQL(embedding_literal))

cursor.execute(query)
results = cursor.fetchall()

for row in results:
    print(row)

cursor.close()
conn.close()
```
